```python
# -*- coding: utf-8 -*-
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf 
import statsmodels.stats.anova as anova
```

```python
# データ読み込み
d = pd.read_csv('data3a.csv')
```

```python
# (1)平均 y.mean のポアソン乱数をd[“y_rnd”]に格納 : 帰無仮説からのデータ生成
d["y_rnd"] = np.random.poisson(d.y.mean(),100)

# (2)d[“y_rnd”]に対する一定モデル、xモデルのglm()の推定結果をfit1,fit2に格納
fit1 = smf.glm(formula="y_rnd ~ 1", data= d, family=sm.families.Poisson()).fit()
fit2 = smf.glm(formula="y_rnd ~ x", data= d, family=sm.families.Poisson()).fit()

# (3)逸脱度の差を計算
print(fit1.deviance - fit2.deviance)
```

# パラメトリックブートストラップ法

```python
# 関数化
def get_dd(d):
    n_sample = len(d.y)
    y_mean = d.y.mean()
    d["y_rnd"] = np.random.poisson(y_mean,n_sample)
    fit1 = smf.glm(formula="y_rnd ~ 1", data= d, family=sm.families.Poisson()).fit()
    fit2 = smf.glm(formula="y_rnd ~ x", data= d, family=sm.families.Poisson()).fit()
    return fit1.deviance - fit2.deviance

def pb(d, n_bootstrap):
    return pd.Series([get_dd(d) for i in range(n_bootstrap)])
```

```python
n_sample = 1000
dd12 = pb(d, n_bootstrap = n_sample)
```

```python
dd12.describe()
```

```python
fig = plt.figure(figsize=(8,4),dpi = 600)
ax = fig.add_subplot(111)
ax.hist(dd12, 100,range=(0,20), ec='black',color = "darkblue")
ax.set_xlabel(r'一定モデルとxモデルの逸脱度の差$\Delta D_{1,2}$')
plt.xlim(0, 20)
plt.ylim(0, 350)
plt.axvline(x= 4.5, linestyle='-',color = "darkred")
plt.axvline(x= 3.81, linestyle=':',color = "darkblue")
plt.show()
```

```python
print("Pvalue:",(dd12 >= 4.5).sum()/1000)
print("P = 0.05 となる逸脱度差の値:",dd12.quantile(0.95))
```

# χ2 分布を使った近似計算法

```python
# fit1 = smf.glm(formula="y ~ 1", data=d, family=sm.families.Poisson()).fit()
# fit2 = smf.glm(formula="y ~ x", data=d, family=sm.families.Poisson()).fit()

# anova.anova_lm(fit1 ,fit2, test = "Chisq") 

n_sample = 1000
chi = pd.Series(np.random.chisquare(1, n_sample))
```

```python
fig = plt.figure(figsize=(8,4),dpi = 600)
ax = fig.add_subplot(111)
ax.hist(chi, 100,range=(0,20), ec='black',color = "darkgreen")
ax.set_xlabel(r'χ2分布による近似')
plt.xlim(0, 20)
plt.ylim(0, 350)
plt.axvline(x= 4.5, linestyle='-',color = "darkred")
plt.axvline(x= 3.81, linestyle=':',color = "darkgreen")
plt.show()
```

```python
print("Pvalue:",(chi >= 4.5).sum()/n_sample)
print("P = 0.05 となる逸脱度の値:",chi.quantile(0.95))
```
